This notebook demonstrates how to use the Data Prep Toolkit (DPK) to convert PDFs and ZIP files to Parquet format.
All required steps are fixed and executable in a standard Jupyter/Colab environment.

```python
# âœ… Import modules
import os
import urllib.request
from dpk_pdf2parquet.transform_python import Pdf2Parquet
```

```python
# âœ… Create input directory and download sample data
os.makedirs("tmp/input", exist_ok=True)
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/data-prep-kit/data-prep-kit/dev/transforms/language/docling2parquet/test-data/input/archive1.zip",
    "tmp/input/archive1.zip"
)
urllib.request.urlretrieve(
    "https://raw.githubusercontent.com/data-prep-kit/data-prep-kit/dev/transforms/language/docling2parquet/test-data/input/redp5110-ch1.pdf",
    "tmp/input/redp5110-ch1.pdf"
)
```

```python
# âœ… Run PDF2Parquet Transformation
Pdf2Parquet(
    input_folder="tmp/input",
    output_folder="tmp/output",
    data_files_to_use=[".pdf", ".zip"]
).transform()
```

```python
# âœ… Load the converted Parquet files
import pyarrow.parquet as pq
import pandas as pd

archive_table = pq.read_table('tmp/output/archive1.parquet')
archive_df = archive_table.to_pandas()
archive_df.head()
```

```python
# âœ… Load second converted file
pdf_table = pq.read_table('tmp/output/redp5110-ch1.parquet')
pdf_df = pdf_table.to_pandas()
pdf_df.head()
```
